```python
def overlap_score(l: list, ion: str) -> float:
    '''
    '''
    l.sort(key=lambda x: len(x))
    
    score = 0
    
    for i, e in enumerate(l[:-1]):
        
        if ion == 'y':
            e = e[::-1]
        
        for j, e2 in enumerate(l[i+1:]):
            
            if ion == 'y':
                e2 = e2[::-1]
            
            score += 1 if e2[:len(e)] == e else -.5
            
    return score
            
```

```python
from collections import defaultdict
from operator import itemgetter

def overlap_assembler(l: list, ion: str) -> str:
    '''
    Take a list of sequences and return the longest sequence that contains
    the most of the rest of the sequences.

    Example:
        l: [ABCDEFG, ABCDE, ABCXY, ABCD, ABC]

        ABCDEFG contains [ABCDE, ABCD, ABC] which is the most of any others
        so ABCDEFG would be returned

    Inputs:
        l:      (list) the sequences to assemble together   
        ion:    (str) the ion this is performed for. y goes right to left, b left to right
    Outputs:
        (str) the sequence that contains the most of the others. If no
                sequence can be assembled, None is returned
    '''
    # sort longest to shortest
    l.sort(key=len, reverse=True)
    
    # got through each element and count the number of smaller 
    # sequences it contains
    assembler = defaultdict(lambda: 0)
    for i, e in enumerate(l[:-1]):
        for e2 in l[i+1:]:
            
            # get left to right or right to left depending on the ion
            cmp_str = e[:len(e2)] if ion == 'b' else e[-len(e2):]
                        
            if e2 == cmp_str:
                assembler[e] += 1
                
    # return the str if possible, None otherwise
    try:    
        return max([(k, v) for k, v in assembler.items()], key=itemgetter(1))[0]
    except:
        return None
```

```python
def lcs(l: list, ion: str) -> str:
    '''
    Return the longest continuous substring of a list of strings

    Inputs:
        l:      (list) sequences to find the lcs of
        ion:    (str) which ion this is for. 'b' goes left to righ, 'y' goes right to left
    Outputs:
        (str) lcs 
    '''

    # if the list has length 1, return that element
    if len(l) == 1:
        return l[0]

    # if the list is emtpy, return empty string
    if len(l) == 0:
        return ''

    # iterate over the range of the shortest string
    for i in range(min(map(len, l))):

        # get the index by the ion type
        idx = i if ion == 'b' else -1 - i

        # if all list elements at the index are equal, continue
        if all([x[idx] == l[0][idx] for x  in l[1:]]):
            continue

        # otherwise return up (not including) this index
        else:
            if ion == 'b':
                return l[0][:idx]
            else:
                return l[0][idx+1:]
```
