```python
def overlap_score(l: list, ion: str) -> float:
    '''
    '''
    l.sort(key=lambda x: len(x))
    
    score = 0
    
    for i, e in enumerate(l[:-1]):
        
        if ion == 'y':
            e = e[::-1]
        
        for j, e2 in enumerate(l[i+1:]):
            
            if ion == 'y':
                e2 = e2[::-1]
            
            score += 1 if e2[:len(e)] == e else -.5
            
    return score
            
```
